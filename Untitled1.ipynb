```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# تحميل البيانات
from google.colab import drive
drive.mount('/content/drive')

file_path = '/content/drive/MyDrive/fraud_detection/fraudData.csv'
df = pd.read_csv(file_path)

print("الأبعاد الأصلية للبيانات:", df.shape)
print("\nالأعمدة الأصلية:")
print(df.columns.tolist())
```

```python
# 1. إنشاء معرف فريد لكل معاملة إذا لم يكن موجوداً
if 'transaction_id' not in df.columns:
    df['transaction_id'] = range(1, len(df) + 1)

# 2. إعادة تسمية الأعمدة لتتوافق مع المطلوب
column_mapping = {
    'customer': 'client_id',
    'merchant': 'account_id',
    'category': 'transaction_type',
    'amount': 'amount',
    'fraud': 'fraud'
}

# تطبيق عملية إعادة التسمية
df.rename(columns=column_mapping, inplace=True)

# 3. معالجة القيم المفقودة
print("\nالقيم المفقودة قبل المعالجة:")
print(df.isnull().sum())

# معالجة القيم المفقودة في الأعمدة الرقمية
numeric_cols = df.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].median(), inplace=True)

# 4. إضافة الأعمدة المطلوبة التي قد تكون غير موجودة
if 'account_type' not in df.columns:
    df['account_type'] = 'standard'  # قيمة افتراضية

if 'office_id' not in df.columns:
    df['office_id'] = 1  # قيمة افتراضية

if 'currency_code' not in df.columns:
    df['currency_code'] = 'EUR'  # قيمة افتراضية

if 'running_balance' not in df.columns:
    # حساب الرصيد الجاري لكل عميل
    df = df.sort_values(['client_id', 'step'])
    df['running_balance'] = df.groupby('client_id')['amount'].cumsum()

# 5. معالجة الطابع الزمني واستخراج ميزات الوقت
if 'step' in df.columns:
    # تحويل الخطوة الزمنية إلى طابع زمني (افترضنا أن كل خطوة تمثل ساعة)
    base_date = datetime(2020, 1, 1)  # تاريخ بداية افتراضي
    df['transaction_datetime'] = df['step'].apply(lambda x: base_date + timedelta(hours=x))

    # استخراج يوم الأسبوع وساعة اليوم
    df['day_of_week'] = df['transaction_datetime'].dt.dayofweek
    df['hour_of_day'] = df['transaction_datetime'].dt.hour

# 6. حساب الوقت منذ آخر معاملة لكل عميل
df = df.sort_values(['client_id', 'transaction_datetime'])
df['time_since_last_txn_mins'] = df.groupby('client_id')['transaction_datetime'].diff().dt.total_seconds() / 60
df['time_since_last_txn_mins'].fillna(0, inplace=True)  # للمعاملات الأولى

# 7. حساب تردد المعاملات خلال آخر 24 ساعة لكل عميل
# ننشئ نافذة زمنية لمدة 24 ساعة
df.set_index('transaction_datetime', inplace=True)
df['txn_frequency_last_24h'] = df.groupby('client_id')['transaction_id'].rolling('24H').count().values
df.reset_index(inplace=True)

# 8. حساب متوسط المبلغ خلال آخر 30 يومًا لكل عميل
# نستخدم نافذة زمنية لمدة 30 يومًا
df.set_index('transaction_datetime', inplace=True)
df['avg_amount_last_30d'] = df.groupby('client_id')['amount'].rolling('30D').mean().values
df.reset_index(inplace=True)

# 9. حساب انحراف المبلغ عن المتوسط (مقياس شذوذ)
df['amount_dev_score'] = (df['amount'] - df['avg_amount_last_30d']) / df['avg_amount_last_30d'].replace(0, 1)
df['amount_dev_score'].replace([np.inf, -np.inf], 0, inplace=True)

# 10. إعادة ترتيب الأعمدة حسب الطلب
desired_columns = [
    'transaction_id', 'client_id', 'account_id', 'account_type', 'office_id',
    'transaction_type', 'amount', 'running_balance', 'currency_code',
    'transaction_datetime', 'day_of_week', 'hour_of_day', 'time_since_last_txn_mins',
    'txn_frequency_last_24h', 'avg_amount_last_30d', 'amount_dev_score', 'fraud'
]

# الاحتفاظ فقط بالأعمدة المتوفرة من القائمة المطلوبة
available_columns = [col for col in desired_columns if col in df.columns]
df = df[available_columns]

# 11. تنظيف أنواع البيانات
if 'transaction_datetime' in df.columns:
    df['transaction_datetime'] = pd.to_datetime(df['transaction_datetime'])

# 12. إزالة أي تكرارات إن وجدت
df.drop_duplicates(inplace=True)

print("\nالأبعاد بعد المعالجة:", df.shape)
print("\nالأعمدة بعد المعالجة:")
print(df.columns.tolist())
print("\nلمحة عن البيانات بعد المعالجة:")
df.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
import gc  # لأداء garbage collection لت freeing الذاكرة
import psutil  # لمراقبة استخدام الذاكرة
import time

# وظيفة لمراقبة استخدام الذاكرة
def monitor_memory():
    process = psutil.Process()
    return process.memory_info().rss / 1024 ** 2  # الذاكرة بالميجابايت

# تحميل البيانات المعالجة مع تقليل استخدام الذاكرة
def load_data_with_memory_efficiency(file_path):
    print("جاري تحميل البيانات مع تحسين استخدام الذاكرة...")
    start_mem = monitor_memory()

    # تحديد أنواع البيانات لتقليل الاستخدام
    dtype_dict = {
        'transaction_id': 'int32',
        'client_id': 'object',  # Changed from 'int32' to 'object'
        'account_id': 'object', # Changed from 'int32' to 'object'
        'account_type': 'category',
        'office_id': 'int8',
        'transaction_type': 'category',
        'amount': 'float32',
        'running_balance': 'float32',
        'currency_code': 'category',
        'day_of_week': 'int8',
        'hour_of_day': 'int8',
        'time_since_last_txn_mins': 'float32',
        'txn_frequency_last_24h': 'float32',
        'avg_amount_last_30d': 'float32',
        'amount_dev_score': 'float32',
        'fraud': 'int8'
    }

    # تحميل البيانات مع أنواع محددة
    df = pd.read_csv(file_path, dtype=dtype_dict)

    # تحويل الأعمدة التاريخية إذا كانت موجودة
    date_columns = ['transaction_datetime']
    for col in date_columns:
        if col in df.columns:
            df[col] = pd.to_datetime(df[col])

    end_mem = monitor_memory()
    print(f"الذاكرة المستخدمة: {end_mem - start_mem:.2f} MB")
    return df

# تحميل البيانات
file_path = '/content/drive/MyDrive/BankSim_data_processed.csv'
df = load_data_with_memory_efficiency(file_path)

# إعداد البيانات للنمذجة مع تقليل استخدام الذاكرة
def prepare_data_memory_efficient(df, target_col='fraud'):
    print("إعداد البيانات بكفاءة ذاكرة...")
    start_mem = monitor_memory()

    # فصل السمات والهدف
    X = df.drop(target_col, axis=1)
    y = df[target_col].values

    # تحديد الأعمدة الرقمية والفئوية
    numeric_cols = X.select_dtypes(include=[np.number]).columns.tolist()
    categorical_cols = X.select_dtypes(include=['category', 'object']).columns.tolist() # Include object for string IDs

    # تحويل الأعمدة الفئوية إلى رقمية (بدون استخدام get_dummies لتقليل الأبعاد)
    for col in categorical_cols:
        # Handle potential non-string values in object columns before cat.codes
        X[col] = X[col].astype(str).str.strip("'") # Remove quotes from IDs
        X[col] = pd.Categorical(X[col]).codes # Convert to numeric codes

    # إزالة الأعمدة غير الضرورية لتقليل الأبعاد
    cols_to_drop = [col for col in X.columns if col not in numeric_cols and
                   col not in ['transaction_datetime']]
    X = X.drop(cols_to_drop, axis=1)

    # تحويل الأعمدة الرقمية إلى أنواع بيانات موفرة للذاكرة
    # Exclude 'transaction_datetime' from astype('float32')
    numeric_cols_to_float = [col for col in X.columns if col in numeric_cols and col != 'transaction_datetime']
    X[numeric_cols_to_float] = X[numeric_cols_to_float].astype('float32')

    # Drop transaction_datetime as it cannot be converted to float
    if 'transaction_datetime' in X.columns:
        X = X.drop(columns=['transaction_datetime'])

    end_mem = monitor_memory()
    print(f"الذاكرة المستخدمة: {end_mem - start_mem:.2f} MB")

    return X, y

# تقسيم البيانات مع تقليل استخدام الذاكرة
def split_data_memory_efficient(X, y, test_size=0.3, random_state=42):
    print("تقسيم البيانات مع الحفاظ على الذاكرة...")
    start_mem = monitor_memory()

    # تقسيم البيانات
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state, stratify=y
    )

    # تحرير الذاكرة المستخدمة بالبيانات الأصلية
    del X, y
    gc.collect()

    end_mem = monitor_memory()
    print(f"الذاكرة المستخدمة: {end_mem - start_mem:.2f} MB")

    return X_train, X_test, y_train, y_test

# تطبيق SMOTE بكفاءة ذاكرة
def apply_smote_memory_efficient(X_train, y_train, sampling_strategy='auto'):
    print("تطبيق SMOTE بكفاءة ذاكرة...")
    start_mem = monitor_memory()

    # تطبيق SMOTE على دفعات للتعامل مع الذاكرة المحدودة
    smote = SMOTE(
        random_state=42,
        sampling_strategy=sampling_strategy,
        k_neighbors=5  # تقليل الجيران لتقليل استخدام الذاكرة
    )

    X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

    end_mem = monitor_memory()
    print(f"الذاكرة المستخدمة: {end_mem - start_mem:.2f} MB")

    return X_resampled, y_resampled

# تطبيق Undersampling بكفاءة ذاكرة
def apply_undersampling_memory_efficient(X_train, y_train, sampling_strategy='auto'):
    print("تطبيق Undersampling بكفاءة ذاكرة...")
    start_mem = monitor_memory()

    rus = RandomUnderSampler(
        random_state=42,
        sampling_strategy=sampling_strategy
    )

    X_resampled, y_resampled = rus.fit_resample(X_train, y_train)

    end_mem = monitor_memory()
    print(f"الذاكرة المستخدمة: {end_mem - start_mem:.2f} MB")

    return X_resampled, y_resampled

# التنفيذ الرئيسي
print("الذاكرة المتاحة قبل المعالجة:", psutil.virtual_memory().available / 1024 ** 2, "MB")

# إعداد البيانات
X, y = prepare_data_memory_efficient(df)

# تقسيم البيانات
X_train, X_test, y_train, y_test = split_data_memory_efficient(X, y)

print(f"توزيع الفئات في بيانات التدريب: الفئة 0: {sum(y_train == 0)}, الفئة 1: {sum(y_train == 1)}")
print(f"توزيع الفئات في بيانات الاختبار: الفئة 0: {sum(y_test == 0)}, الفئة 1: {sum(y_test == 1)}")

# تطبيق SMOTE (اختياري - استخدم واحدة فقط حسب احتياجاتك)
try:
    X_train_smote, y_train_smote = apply_smote_memory_efficient(X_train, y_train)
    print(f"بعد SMOTE - الفئة 0: {sum(y_train_smote == 0)}, الفئة 1: {sum(y_train_smote == 1)}")

    # حفظ النتائج
    np.save('/content/drive/MyDrive/X_train_smote.npy', X_train_smote)
    np.save('/content/drive/MyDrive/y_train_smote.npy', y_train_smote)

    # تحرير الذاكرة
    del X_train_smote, y_train_smote
    gc.collect()
except MemoryError:
    print("لا توجد ذاكرة كافية لتطبيق SMOTE، جاري تطبيق Undersampling بدلاً من ذلك")

# تطبيق Undersampling
X_train_under, y_train_under = apply_undersampling_memory_efficient(X_train, y_train)
print(f"بعد Undersampling - الفئة 0: {sum(y_train_under == 0)}, الفئة 1: {sum(y_train_under == 1)}")

# حفظ النتائج
np.save('/content/drive/MyDrive/X_train_under.npy', X_train_under)
np.save('/content/drive/MyDrive/y_train_under.npy', y_train_under)
np.save('/content/drive/MyDrive/X_test.npy', X_test)
np.save('/content/drive/MyDrive/y_test.npy', y_test)

print("تم الانتهاء من المعالجة وحفظ النتائج!")
```

```python
import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, precision_recall_curve, auc, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns
import gc
import joblib
import time

# تحميل البيانات المعالجة
print("جاري تحميل البيانات المعالجة...")
X_train_under = np.load('/content/drive/MyDrive/X_train_under.npy', allow_pickle=True)
y_train_under = np.load('/content/drive/MyDrive/y_train_under.npy', allow_pickle=True)
X_test = np.load('/content/drive/MyDrive/X_test.npy', allow_pickle=True)
y_test = np.load('/content/drive/MyDrive/y_test.npy', allow_pickle=True)

# تحويل البيانات إلى تنسيق LightGBM الأمثل
print("تحضير البيانات لـ LightGBM...")
train_data = lgb.Dataset(X_train_under, label=y_train_under)
test_data = lgb.Dataset(X_test, label=y_test, reference=train_data)

# معلمات LightGBM المثلى لاكتشاف الاحتيال
params = {
    'boosting_type': 'gbdt',
    'objective': 'binary',
    'metric': ['auc', 'binary_logloss'],
    'num_leaves': 31,
    'learning_rate': 0.05,
    'feature_fraction': 0.8,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'verbose': 0,
    'random_state': 42,
    'n_jobs': 2,  # تقليل عدد الوظائف لتوفير الذاكرة
    'max_depth': 7,  # تقليل العمق لتقليل استخدام الذاكرة
    'min_data_in_leaf': 20,  # زيادة الحد الأدنى للبيانات في الورقة
    'is_unbalance': True  # معالجة عدم التوازن في البيانات
}

# تدريب نموذج LightGBM
print("بدء تدريب نموذج LightGBM...")
start_time = time.time()

model = lgb.train(
    params,
    train_data,
    num_boost_round=1000,
    valid_sets=[test_data],
    # early_stopping_rounds=50, # Removed this line
    # verbose_eval=50 # Removed this line
)

end_time = time.time()
print(f"تم الانتهاء من التدريب في {end_time - start_time:.2f} ثانية")

# عمل تنبؤات
print("جاري عمل التنبؤات...")
y_pred_proba = model.predict(X_test, num_iteration=model.best_iteration)
y_pred = (y_pred_proba >= 0.5).astype(int)

# تقييم النموذج
print("\n" + "="*50)
print("تقييم أداء النموذج")
print("="*50)

# تقرير التصنيف
print("\nتقرير التصنيف:")
print(classification_report(y_test, y_pred))

# مصفوفة الارتباك
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('مصفوفة الارتباك')
plt.ylabel('التصنيف الحقيقي')
plt.xlabel('التصنيف المتوقع')
plt.show()

# حساب AUC-ROC
roc_auc = roc_auc_score(y_test, y_pred_proba)
print(f"منطقة تحت منحنى ROC (AUC): {roc_auc:.4f}")

# منحنى ROC
plt.figure(figsize=(8, 6))
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
plt.plot(fpr, tpr, label=f'LightGBM (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('معدل الإيجابيات الكاذبة')
plt.ylabel('معدل الإيجابيات الحقيقية')
plt.title('منحنى ROC')
plt.legend(loc='lower right')
plt.show()

# منحنى Precision-Recall
precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)
pr_auc = auc(recall, precision)
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, label=f'LightGBM (AUC = {pr_auc:.4f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('منحنى Precision-Recall')
plt.legend(loc='lower left')
plt.show()

# أهمية الميزات
print("\nأهم 10 ميزات في النموذج:")
feature_importance = pd.DataFrame({
    'feature': [f'Feature_{i}' for i in range(len(model.feature_importance()))],
    'importance': model.feature_importance()
}).sort_values('importance', ascending=False)

plt.figure(figsize=(10, 8))
sns.barplot(x='importance', y='feature', data=feature_importance.head(10))
plt.title('أهم 10 ميزات في نموذج LightGBM')
plt.show()

print(feature_importance.head(10))

# حفظ النموذج
print("\nجاري حفظ النموذج...")
joblib.dump(model, '/content/drive/MyDrive/lightgbm_fraud_model.pkl')
print("تم حفظ النموذج بنجاح!")

# تحرير الذاكرة
del X_train_under, y_train_under, train_data
gc.collect()

# تقييم أداء النموذج على البيانات غير المتوازنة الأصلية (اختياري)
print("\nتقييم إضافي: أداء النموذج مع عتبات تصنيف مختلفة")

# تجربة عتبات مختلفة للتصنيف
thresholds = [0.3, 0.4, 0.5, 0.6, 0.7]
results = []

for threshold in thresholds:
    y_pred_thresh = (y_pred_proba >= threshold).astype(int)
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred_thresh).ravel()

    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    results.append({
        'threshold': threshold,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'false_positives': fp,
        'false_negatives': fn
    })

results_df = pd.DataFrame(results)
print("\nأداء النموذج مع عتبات تصنيف مختلفة:")
print(results_df)

# العثور على أفضل عتبة بناء على F1-score
best_threshold = results_df.loc[results_df['f1'].idxmax()]
print(f"\nأفضل عتبة بناء على F1-score: {best_threshold['threshold']}")
print(f"الدقة: {best_threshold['precision']:.4f}, الاستدعاء: {best_threshold['recall']:.4f}, F1: {best_threshold['f1']:.4f}")

# استخدام أفضل عتبة للتنبؤات النهائية
best_threshold_value = best_threshold['threshold']
y_pred_final = (y_pred_proba >= best_threshold_value).astype(int)

print("\nالتقرير النهائي مع أفضل عتبة:")
print(classification_report(y_test, y_pred_final))

# حفظ التنبؤات النهائية
final_results = pd.DataFrame({
    'actual': y_test,
    'predicted_probability': y_pred_proba,
    'predicted_class': y_pred_final
})

final_results.to_csv('/content/drive/MyDrive/fraud_predictions.csv', index=False)
print("تم حفظ التنبؤات النهائية!")
```

```python
import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import (classification_report, confusion_matrix, roc_auc_score,
                             precision_recall_curve, auc, average_precision_score, f1_score)
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import time
import gc
import warnings
warnings.filterwarnings('ignore')

# 1. تحميل البيانات المعالجة
print("📂 جاري تحميل البيانات المعالجة...")
df = pd.read_csv('/content/drive/MyDrive/BankSim_data_processed.csv')

# إذا كان هناك عمود تاريخ، تحويله إلى نوع تاريخ
if 'transaction_datetime' in df.columns:
    df['transaction_datetime'] = pd.to_datetime(df['transaction_datetime'])

# عرض معلومات عن البيانات
print(f"📊 شكل البيانات: {df.shape}")
print(f"📈 توزيع الفئات:\n{df['fraud'].value_counts()}")
print(f"📊 نسبة الاحتيال: {df['fraud'].mean()*100:.2f}%")

# 2. فصل السمات والهدف
X = df.drop('fraud', axis=1)
y = df['fraud']

# إزالة عمود التاريخ قبل تطبيق SMOTE
if 'transaction_datetime' in X.columns:
    X = X.drop('transaction_datetime', axis=1)

# 3. تقسيم البيانات مع الحفاظ على توزيع الفئات
print("✂️ تقسيم البيانات إلى تدريب واختبار...")
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"📊 توزيع الفئات في التدريب: {pd.Series(y_train).value_counts().to_dict()}")
print(f"📊 توزيع الفئات في الاختبار: {pd.Series(y_test).value_counts().to_dict()}")

# 4. تحضير البيانات للتدريب
# تحويل البيانات الفئوية إلى رقمية
categorical_cols = X_train.select_dtypes(include=['object', 'category']).columns.tolist()
for col in categorical_cols:
    if col in X_train.columns:
        X_train[col] = X_train[col].astype('category').cat.codes
        X_test[col] = X_test[col].astype('category').cat.codes


# 5. تطبيق تقنيات موازنة البيانات (فقط على بيانات التدريب)
print("⚖️ تطبيق SMOTE على بيانات التدريب فقط...")

# حساب نسبة الموازنة المثلى (نسبة الفئة الأقل إلى الفئة الأكبر)
fraud_ratio = sum(y_train == 1) / sum(y_train == 0)
sampling_strategy = {1: int(sum(y_train == 0) * 0.3), 0: sum(y_train == 0)}  # نسبة 30% احتيال

# تطبيق SMOTE
smote = SMOTE(random_state=42, sampling_strategy=sampling_strategy, k_neighbors=5)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

print(f"📊 توزيع الفئات بعد SMOTE: {pd.Series(y_train_res).value_counts().to_dict()}")

# 6. ضبط معلمات LightGBM بشكل متقدم
print("⚙️ ضبط معلمات LightGBم بشكل متقدم...")

# حساب وزن الفئة للفئة الاحتيالية
scale_pos_weight = len(y_train_res[y_train_res==0]) / len(y_train_res[y_train_res==1])

params_optimized = {
    'boosting_type': 'gbdt',
    'objective': 'binary',
    'metric': ['auc', 'binary_logloss', 'average_precision'],
    'num_leaves': 63,
    'learning_rate': 0.05,
    'feature_fraction': 0.8,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'verbose': -1,
    'random_state': 42,
    'n_jobs': 2,
    'max_depth': 7,
    'min_data_in_leaf': 50,
    'lambda_l1': 1.0,
    'lambda_l2': 1.0,
    'min_gain_to_split': 0.01,
    'scale_pos_weight': scale_pos_weight,
    'subsample_for_bin': 200000,
    'subsample': 0.8,
    'subsample_freq': 1,
    'colsample_bytree': 0.8,
    'reg_alpha': 0.5,
    'reg_lambda': 0.5
}

# 7. تدريب النموذج مع التحقق من الأداء
print("🔧 بدء تدريب النموذج...")
start_time = time.time()

# تحويل البيانات إلى تنسيق LightGBM
train_data = lgb.Dataset(X_train_res, label=y_train_res)
test_data = lgb.Dataset(X_test, label=y_test, reference=train_data)

# التدريب مع التوقف المبكر باستخدام callbacks
model = lgb.train(
    params_optimized,
    train_data,
    num_boost_round=1000,
    valid_sets=[test_data],
    callbacks=[
        lgb.early_stopping(stopping_rounds=50, verbose=True, first_metric_only=False),
        lgb.log_evaluation(period=100),
        lgb.record_evaluation({})
    ]
)

end_time = time.time()
print(f"⏱️ تم الانتهاء من التدريب في {end_time - start_time:.2f} ثانية")
print(f"📊 أفضل عدد للتكرارات: {model.best_iteration}")

# 8. عمل التنبؤات وتقييم النموذج
print("🔮 جاري عمل التنبؤات...")
y_pred_proba = model.predict(X_test, num_iteration=model.best_iteration)

# العثور على أفضل عتبة باستخدام منحنى Precision-Recall
precision, recall, thresholds = precision_recall_curve(y_test, y_pred_proba)
f1_scores = 2 * (precision[:-1] * recall[:-1]) / (precision[:-1] + recall[:-1] + 1e-9)
best_threshold = thresholds[np.argmax(f1_scores)]

print(f"🎯 أفضل عتبة بناء على F1-score: {best_threshold:.4f}")

# تطبيق أفضل عتبة
y_pred = (y_pred_proba >= best_threshold).astype(int)

# 9. تقييم النموذج بشكل شامل
print("\n" + "="*60)
print("📊 تقييم أداء النموذج بعد التحسين")
print("="*60)

# تقرير التصنيف
print("\n📝 تقرير التصنيف:")
print(classification_report(y_test, y_pred, target_names=['شرعي', 'احتيال']))

# مصفوفة الارتباك
plt.figure(figsize=(10, 8))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['شرعي', 'احتيال'],
            yticklabels=['شرعي', 'احتيال'])
plt.title('مصفوفة الارتباك')
plt.ylabel('التصنيف الحقيقي')
plt.xlabel('التصنيف المتوقع')
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/confusion_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

# حساب المقاييس الإضافية
roc_auc = roc_auc_score(y_test, y_pred_proba)
avg_precision = average_precision_score(y_test, y_pred_proba)
f1 = f1_score(y_test, y_pred)

print(f"📈 منطقة تحت منحنى ROC (AUC): {roc_auc:.4f}")
print(f"📊 متوسط الدقة (Average Precision): {avg_precision:.4f}")
print(f"⚖️ F1-Score: {f1:.4f}")

# منحنى ROC
plt.figure(figsize=(10, 8))
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
plt.plot(fpr, tpr, label=f'LightGBM (AUC = {roc_auc:.4f})', linewidth=2)
plt.plot([0, 1], [0, 1], 'k--', linewidth=1)
plt.xlabel('معدل الإيجابيات الكاذبة')
plt.ylabel('معدل الإيجابيات الحقيقية')
plt.title('منحنى ROC')
plt.legend(loc='lower right')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/roc_curve.png', dpi=300, bbox_inches='tight')
plt.show()

# منحنى Precision-Recall
plt.figure(figsize=(10, 8))
precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)
pr_auc = auc(recall, precision)
plt.plot(recall, precision, label=f'LightGBM (AP = {avg_precision:.4f})', linewidth=2)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('منحنى Precision-Recall')
plt.legend(loc='lower left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/precision_recall_curve.png', dpi=300, bbox_inches='tight')
plt.show()

# 10. أهمية الميزات
print("\n🔍 أهم 15 ميزة في النموذج:")
feature_importance = pd.DataFrame({
    'feature': X_train_res.columns,
    'importance': model.feature_importance(importance_type='gain')
}).sort_values('importance', ascending=False)


plt.figure(figsize=(12, 10))
sns.barplot(x='importance', y='feature', data=feature_importance.head(15))
plt.title('أهم 15 ميزة في نموذج LightGBM')
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/feature_importance.png', dpi=300, bbox_inches='tight')
plt.show()

print(feature_importance.head(15))

# 11. التحقق من عدم وجود overfitting
print("\n🔎 التحقق من Overfitting:")
train_pred_proba = model.predict(X_train_res, num_iteration=model.best_iteration)
train_pred = (train_pred_proba >= best_threshold).astype(int)

print("أداء النموذج على بيانات التدريب:")
print(classification_report(y_train_res, train_pred, target_names=['شرعي', 'احتيال']))

print("أداء النموذج على بيانات الاختبار:")
print(classification_report(y_test, y_pred, target_names=['شرعي', 'احتيال']))

# 12. تحليل الأداء مع عتبات مختلفة
print("\n📊 تحليل الأداء مع عتبات مختلفة:")
thresholds_to_try = np.arange(0.1, 0.9, 0.1)
results = []

for threshold in thresholds_to_try:
    y_pred_thresh = (y_pred_proba >= threshold).astype(int)
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred_thresh).ravel()

    precision_val = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall_val = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1_val = 2 * (precision_val * recall_val) / (precision_val + recall_val + 1e-9)

    results.append({
        'threshold': threshold,
        'precision': precision_val,
        'recall': recall_val,
        'f1': f1_val,
        'false_positives': fp,
        'false_negatives': fn,
        'true_positives': tp
    })

results_df = pd.DataFrame(results)
print(results_df.round(4))

# رسم منحنى Precision-Recall مقابل العتبات
plt.figure(figsize=(12, 8))
plt.plot(results_df['threshold'], results_df['precision'], label='Precision', marker='o', linewidth=2)
plt.plot(results_df['threshold'], results_df['recall'], label='Recall', marker='s', linewidth=2)
plt.plot(results_df['threshold'], results_df['f1'], label='F1-Score', marker='^', linewidth=2)
plt.xlabel('عتبة التصنيف')
plt.ylabel('القيمة')
plt.title('تأثير عتبة التصنيف على Precision و Recall و F1-Score')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/threshold_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

# 13. تقييم نهائي مع أفضل عتبة
best_idx = results_df['f1'].idxmax()
best_result = results_df.loc[best_idx]

print(f"\n🎉 أفضل أداء عند العتبة: {best_result['threshold']:.2f}")
print(f"📊 الدقة: {best_result['precision']:.4f}")
print(f"📈 الاستدعاء: {best_result['recall']:.4f}")
print(f"⚖️ F1-Score: {best_result['f1']:.4f}")
print(f"✅ الإيجابيات الحقيقية: {best_result['true_positives']}")
print(f"❌ الإيجابيات الكاذبة: {best_result['false_positives']}")
print(f"⚠️ السلبيات الكاذبة: {best_result['false_negatives']}")

# 14. حفظ النموذج والنتائج
print("\n💾 جاري حفظ النموذج والنتائج...")
joblib.dump(model, '/content/drive/MyDrive/optimized_lightgbm_fraud_model.pkl')

# حفظ التنبؤات والنتائج
results_df = pd.DataFrame({
    'actual': y_test,
    'predicted_probability': y_pred_proba,
    'predicted_class': y_pred
})
results_df.to_csv('/content/drive/MyDrive/fraud_detection_results.csv', index=False)

# حفظ معلمات النموذج
params_df = pd.DataFrame.from_dict(params_optimized, orient='index', columns=['value'])
params_df.to_csv('/content/drive/MyDrive/model_parameters.csv')

print("✅ تم حفظ النموذج والنتائج بنجاح!")

# 15. تحليل القيمة الاقتصادية للنموذج
print("\n💰 تحليل القيمة الاقتصادية للنموذج:")

# افترض أن متوسط قيمة المعاملة الاحتيالية هو 500 دولار
avg_fraud_amount = 500
total_frauds = sum(y_test)
detected_frauds = best_result['true_positives']
missed_frauds = best_result['false_negatives']
false_positives = best_result['false_positives']

# تكلفة التحقيق في كل إنذار كاذب (افترض 10 دولارات)
cost_per_false_positive = 10

# القيمة المحفوظة من اكتشاف الاحتيال
saved_value = detected_frauds * avg_fraud_amount

# تكلفة الإنذارات الكاذبة
false_positive_cost = false_positives * cost_per_false_positive

# القيمة الصافية
net_value = saved_value - false_positive_cost

print(f"💵 القيمة المحفوظة من اكتشاف الاحتيال: ${saved_value}")
print(f"💸 تكلفة الإنذارات الكاذبة: ${false_positive_cost}")
print(f"💰 القيمة الصافية: ${net_value}")
print(f"📊 نسبة الاكتشاف: {detected_frauds/total_frauds*100:.2f}%")

# تحرير الذاكرة
# del X_train, X_test, y_train, y_test, X_train_res, y_train_res # Removed this line
gc.collect()

print("\n🎉 تم الانتهاء من عملية التحسين بنجاح!")
```

```python
import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, precision_recall_curve, auc, average_precision_score, f1_score, precision_score, recall_score
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import json
import time
import gc
import warnings
warnings.filterwarnings('ignore')

# 1. حفظ كل الإعدادات والمعلمات
def save_model_artifacts(model, X_test, y_test, y_pred_proba, y_pred, params_optimized, best_threshold):
    """
    حفظ كل مكونات النموذج والنتائج
    """
    print("💾 بدء عملية الحفظ...")

    # إنشاء مجلد للحفظ إذا لم يكن موجوداً
    import os
    save_dir = '/content/drive/MyDrive/fraud_detection_model'
    os.makedirs(save_dir, exist_ok=True)

    # أ- حفظ النموذج نفسه
    model_path = f'{save_dir}/lightgbm_fraud_model.pkl'
    joblib.dump(model, model_path)
    print(f"✅ النموذج محفوظ في: {model_path}")

    # ب- حفظ معلمات النموذج
    params_path = f'{save_dir}/model_parameters.json'
    with open(params_path, 'w') as f:
        json.dump(params_optimized, f, indent=4)
    print(f"✅ معلمات النموذج محفوظة في: {params_path}")

    # ج- حفظ التنبؤات والنتائج
    results_df = pd.DataFrame({
        'actual': y_test,
        'predicted_probability': y_pred_proba,
        'predicted_class': y_pred,
        'correct_prediction': (y_test == y_pred)
    })

    results_path = f'{save_dir}/prediction_results.csv'
    results_df.to_csv(results_path, index=False)
    print(f"✅ نتائج التنبؤات محفوظة في: {results_path}")

    # د- حفظ إحصاءات الأداء
    performance_report = classification_report(y_test, y_pred, output_dict=True)
    performance_path = f'{save_dir}/performance_metrics.json'
    with open(performance_path, 'w') as f:
        json.dump(performance_report, f, indent=4)
    print(f"✅ تقرير الأداء محفوظ في: {performance_path}")

    # ه- حفظ عتبة التصنيف المثلى
    threshold_data = {
        'best_threshold': float(best_threshold),
        'f1_score': float(f1_score(y_test, y_pred)),
        'roc_auc': float(roc_auc_score(y_test, y_pred_proba)),
        'average_precision': float(average_precision_score(y_test, y_pred_proba))
    }

    threshold_path = f'{save_dir}/optimal_threshold.json'
    with open(threshold_path, 'w') as f:
        json.dump(threshold_data, f, indent=4)
    print(f"✅ عتبة التصنيف المثلى محفوظة في: {threshold_path}")

    # و- حفظ مصفوفة الارتباك
    cm = confusion_matrix(y_test, y_pred)
    cm_df = pd.DataFrame(cm,
                        index=['حقيقي - شرعي', 'حقيقي - احتيال'],
                        columns=['متوقع - شرعي', 'متوقع - احتيال'])
    cm_path = f'{save_dir}/confusion_matrix.csv'
    cm_df.to_csv(cm_path)
    print(f"✅ مصفوفة الارتباك محفوظة في: {cm_path}")

    return save_dir

# 2. حفظ الرسوم البيانية
def save_plots(model, X_test, y_test, y_pred_proba, save_dir):
    """
    حفظ جميع الرسوم البيانية
    """
    print("📊 حفظ الرسوم البيانية...")

    # أ- منحنى ROC
    plt.figure(figsize=(10, 8))
    fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
    roc_auc = roc_auc_score(y_test, y_pred_proba)
    plt.plot(fpr, tpr, label=f'LightGBM (AUC = {roc_auc:.4f})', linewidth=2)
    plt.plot([0, 1], [0, 1], 'k--', linewidth=1)
    plt.xlabel('معدل الإيجابيات الكاذبة')
    plt.ylabel('معدل الإيجابيات الحقيقية')
    plt.title('منحنى ROC')
    plt.legend(loc='lower right')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(f'{save_dir}/roc_curve.png', dpi=300, bbox_inches='tight')
    plt.close()

    # ب- منحنى Precision-Recall
    plt.figure(figsize=(10, 8))
    precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)
    avg_precision = average_precision_score(y_test, y_pred_proba)
    plt.plot(recall, precision, label=f'LightGBM (AP = {avg_precision:.4f})', linewidth=2)
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('منحنى Precision-Recall')
    plt.legend(loc='lower left')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(f'{save_dir}/precision_recall_curve.png', dpi=300, bbox_inches='tight')
    plt.close()

    # ج- مصفوفة الارتباك
    plt.figure(figsize=(10, 8))
    cm = confusion_matrix(y_test, (y_pred_proba >= 0.5).astype(int))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['شرعي', 'احتيال'],
                yticklabels=['شرعي', 'احتيال'])
    plt.title('مصفوفة الارتباك')
    plt.ylabel('التصنيف الحقيقي')
    plt.xlabel('التصنيف المتوقع')
    plt.tight_layout()
    plt.savefig(f'{save_dir}/confusion_matrix_plot.png', dpi=300, bbox_inches='tight')
    plt.close()

    # د- أهمية الميزات
    feature_importance = pd.DataFrame({
        'feature': X_test.columns,
        'importance': model.feature_importance(importance_type='gain')
    }).sort_values('importance', ascending=False)

    plt.figure(figsize=(12, 10))
    sns.barplot(x='importance', y='feature', data=feature_importance.head(15))
    plt.title('أهم 15 ميزة في نموذج LightGBM')
    plt.tight_layout()
    plt.savefig(f'{save_dir}/feature_importance.png', dpi=300, bbox_inches='tight')
    plt.close()

    # حفظ بيانات أهمية الميزات
    feature_importance.to_csv(f'{save_dir}/feature_importance.csv', index=False)

    print("✅ جميع الرسوم البيانية محفوظة")

# 3. حفظ تحليل العتبات
def save_threshold_analysis(y_test, y_pred_proba, save_dir):
    """
    حفظ تحليل تأثير عتبات التصنيف المختلفة
    """
    print("📈 حفظ تحليل العتبات...")

    thresholds_to_try = np.arange(0.1, 0.9, 0.05)
    results = []

    for threshold in thresholds_to_try:
        y_pred_thresh = (y_pred_proba >= threshold).astype(int)
        tn, fp, fn, tp = confusion_matrix(y_test, y_pred_thresh).ravel()

        precision_val = tp / (tp + fp) if (tp + fp) > 0 else 0
        recall_val = tp / (tp + fn) if (tp + fn) > 0 else 0
        f1_val = 2 * (precision_val * recall_val) / (precision_val + recall_val + 1e-9)

        results.append({
            'threshold': threshold,
            'precision': precision_val,
            'recall': recall_val,
            'f1_score': f1_val,
            'true_positives': tp,
            'false_positives': fp,
            'false_negatives': fn,
            'true_negatives': tn
        })

    threshold_df = pd.DataFrame(results)
    threshold_path = f'{save_dir}/threshold_analysis.csv'
    threshold_df.to_csv(threshold_path, index=False)

    # رسم منحنى العتبات
    plt.figure(figsize=(12, 8))
    plt.plot(threshold_df['threshold'], threshold_df['precision'], label='Precision', marker='o', linewidth=2)
    plt.plot(threshold_df['threshold'], threshold_df['recall'], label='Recall', marker='s', linewidth=2)
    plt.plot(threshold_df['threshold'], threshold_df['f1_score'], label='F1-Score', marker='^', linewidth=2)
    plt.xlabel('عتبة التصنيف')
    plt.ylabel('القيمة')
    plt.title('تأثير عتبة التصنيف على الأداء')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(f'{save_dir}/threshold_analysis_plot.png', dpi=300, bbox_inches='tight')
    plt.close()

    print(f"✅ تحليل العتبات محفوظ في: {threshold_path}")

# 4. حفظ تقرير مفصل عن النموذج
def save_model_report(model, X_train_res, y_train_res, X_test, y_test, y_pred, y_pred_proba, best_threshold, params_optimized, save_dir):
    """
    حفظ تقرير مفصل عن أداء النموذج
    """
    print("📝 حفظ التقرير المفصل...")

    # حساب المقاييس
    roc_auc = roc_auc_score(y_test, y_pred_proba)
    avg_precision = average_precision_score(y_test, y_pred_proba)
    f1 = f1_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)

    # إنشاء التقرير
    report = {
        'model_info': {
            'model_type': 'LightGBM',
            'boosting_type': params_optimized.get('boosting_type', 'gbdt'),
            'num_leaves': params_optimized.get('num_leaves', 31),
            'learning_rate': params_optimized.get('learning_rate', 0.05),
            'training_time': 'تم حساب الوقت أثناء التدريب' # يمكنك إضافة الوقت الفعلي إذا كان متاحاً
        },
        'dataset_info': {
            'training_samples': len(X_train_res),
            'test_samples': len(X_test),
            'fraud_percentage_training': f"{np.mean(y_train_res) * 100:.2f}%",
            'fraud_percentage_test': f"{np.mean(y_test) * 100:.2f}%",
            'number_of_features': X_train_res.shape[1]
        },
        'performance_metrics': {
            'roc_auc': roc_auc,
            'average_precision': avg_precision,
            'f1_score': f1,
            'precision': precision,
            'recall': recall,
            'best_threshold': float(best_threshold) # التأكد من أنه float للحفظ في JSON
        },
        'confusion_matrix': {
            'true_positives': int(confusion_matrix(y_test, y_pred)[1, 1]),
            'false_positives': int(confusion_matrix(y_test, y_pred)[0, 1]),
            'false_negatives': int(confusion_matrix(y_test, y_pred)[1, 0]),
            'true_negatives': int(confusion_matrix(y_test, y_pred)[0, 0])
        },
        'training_parameters': params_optimized
    }

    # حفظ التقرير
    report_path = f'{save_dir}/model_report.json'
    with open(report_path, 'w', encoding='utf-8') as f:
        json.dump(report, f, indent=4, ensure_ascii=False)

    print(f"✅ التقرير المفصل محفوظ في: {report_path}")

# 5. الدالة الرئيسية للحفظ
def save_complete_workflow(model, X_train_res, y_train_res, X_test, y_test, y_pred_proba, y_pred, params_optimized, best_threshold):
    """
    حفظ كامل سير العمل والنتائج
    """
    # حفظ كل المكونات
    save_dir = save_model_artifacts(model, X_test, y_test, y_pred_proba, y_pred, params_optimized, best_threshold)
    save_plots(model, X_test, y_test, y_pred_proba, save_dir)
    save_threshold_analysis(y_test, y_pred_proba, save_dir)
    save_model_report(model, X_train_res, y_train_res, X_test, y_test, y_pred, y_pred_proba, best_threshold, params_optimized, save_dir)

    # إنشاء ملف README تلقائي
    # تأكد من أن المتغيرات المستخدمة هنا متاحة
    try:
        readme_content = f"""
        # نموذج كشف الاحتيال - حفظ كامل العمل

        ## 📁 محتويات المجلد
        - `lightgbm_fraud_model.pkl`: النموذج المدرب
        - `model_parameters.json`: معلمات النموذج
        - `prediction_results.csv`: نتائج التنبؤات
        - `performance_metrics.json`: مقاييس الأداء
        - `optimal_threshold.json`: عتبة التصنيف المثلى
        - `confusion_matrix.csv`: مصفوفة الارتباك
        - `feature_importance.csv`: أهمية الميزات
        - `threshold_analysis.csv`: تحليل العتبات
        - `model_report.json`: التقرير المفصل

        ## 📊 الرسوم البيانية
        - `roc_curve.png`: منحنى ROC
        - `precision_recall_curve.png`: منحنى Precision-Recall
        - `confusion_matrix_plot.png`: مصفوفة الارتباك (رسم)
        - `feature_importance.png`: أهمية الميزات
        - `threshold_analysis_plot.png`: تحليل العتبات

        ## 🎯 أداء النموذج
        - **AUC-ROC**: {roc_auc_score(y_test, y_pred_proba):.4f}
        - **Average Precision**: {average_precision_score(y_test, y_pred_proba):.4f}
        - **F1-Score**: {f1_score(y_test, y_pred):.4f}
        - **العتبة المثلى**: {float(best_threshold):.4f}

        ## 📅 تاريخ الحفظ
        {time.strftime('%Y-%m-%d %H:%M:%S')}
        """

        with open(f'{save_dir}/README.md', 'w', encoding='utf-8') as f:
            f.write(readme_content)
        print("✅ تم إنشاء ملف README.md")

    except Exception as e:
        print(f"⚠️ حدث خطأ أثناء إنشاء ملف README.md: {e}")


    print(f"\n🎉 تم حفظ كل العمل بنجاح في: {save_dir}")
    print("📋 يمكنك الآن استعادة النموذج والنتائج في أي وقت")

    return save_dir

# 6. دالة لتحميل النموذج لاحقاً
def load_saved_model(save_dir='/content/drive/MyDrive/fraud_detection_model'):
    """
    تحميل النموذج والنتائج المحفوظة
    """
    print("🔍 جاري تحميل النموذج المحفوظ...")

    # تحميل النموذج
    model_path = f'{save_dir}/lightgbm_fraud_model.pkl'
    model = joblib.load(model_path)

    # تحميل المعلمات
    params_path = f'{save_dir}/model_parameters.json'
    with open(params_path, 'r') as f:
        params = json.load(f)

    # تحميل العتبة المثلى
    threshold_path = f'{save_dir}/optimal_threshold.json'
    with open(threshold_path, 'r') as f:
        threshold_data = json.load(f)

    # تحميل تقرير الأداء
    report_path = f'{save_dir}/model_report.json'
    with open(report_path, 'r', encoding='utf-8') as f:
        report = json.load(f)

    print("✅ تم تحميل النموذج والبيانات بنجاح")
    return model, params, threshold_data, report

# 7. دالة للتنبؤ باستخدام النموذج المحفوظ
def predict_with_saved_model(new_data, save_dir='/content/drive/MyDrive/fraud_detection_model', threshold=None):
    """
    استخدام النموذج المحفوظ للتنبؤ ببيانات جديدة
    """
    # تحميل النموذج والمعلمات
    model, params, threshold_data, report = load_saved_model(save_dir)

    # استخدام العتبة المثلى إذا لم يتم تحديد عتبة
    if threshold is None:
        threshold = threshold_data['best_threshold']

    # تحضير البيانات الجديدة (نفس معالجة بيانات التدريب)
    # تحتاج للتأكد من تطبيق نفس خطوات المعالجة المسبقة على البيانات الجديدة
    # بما في ذلك تحويل الأعمدة الفئوية وإزالة عمود التاريخ إذا كان موجوداً
    categorical_cols = new_data.select_dtypes(include=['object', 'category']).columns.tolist()
    for col in categorical_cols:
        if col in new_data.columns:
            new_data[col] = new_data[col].astype('category').cat.codes

    cols_to_drop_predict = ['transaction_datetime'] if 'transaction_datetime' in new_data.columns else []
    new_data = new_data.drop(columns=cols_to_drop_predict, errors='ignore')


    # التنبؤ
    predictions = model.predict(new_data)
    predicted_classes = (predictions >= threshold).astype(int)

    # إضافة التنبؤات إلى البيانات
    new_data['fraud_probability'] = predictions
    new_data['predicted_fraud'] = predicted_classes

    print(f"✅ تم التنبؤ بـ {sum(predicted_classes)} معاملة احتيالية محتملة")

    return new_data

# 8. التنفيذ الرئيسي - حفظ كل العمل
print("🚀 بدء عملية حفظ كامل العمل...")

# استبدل هذه المتغيرات بالقيم الفعلية من كودك السابق
# model = نموذجك المدرب
# X_train_res, y_train_res = بيانات التدريب بعد المعالجة
# X_test, y_test = بيانات الاختبار
# y_pred_proba = احتمالات التنبؤ
# y_pred = التصنيفات النهائية
# params_optimized = معلمات النموذج
# best_threshold = أفضل عتبة

# قم باستدعاء الدالة الرئيسية للحفظ
save_dir = save_complete_workflow(
    model, X_train_res, y_train_res, X_test, y_test,
    y_pred_proba, y_pred, params_optimized, best_threshold
)

print("\n" + "="*60)
print("💾 عملية الحفظ الكاملة مكتملة!")
print("="*60)
print(f"📁 جميع الملفات محفوظة في: {save_dir}")
print("📋 يمكنك الآن:")
print("   - استعادة النموذج باستخدام load_saved_model()")
print("   - التنبؤ ببيانات جديدة باستخدام predict_with_saved_model()")
print("   - مراجعة التقارير والرسوم البيانية المحفوظة")
```
